```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_1940 = pd.read_csv('data_1940.csv') 
df_1950 = pd.read_csv('data_1950.csv') 
df_1960 = pd.read_csv('data_1960.csv') 
df_1970 = pd.read_csv('data_1970.csv') 
df_1980 = pd.read_csv('data_1980.csv') 
df_1990 = pd.read_csv('data_1990.csv') 
df_2000 = pd.read_csv('data_2000.csv') 
df_2010 = pd.read_csv('data_2010.csv') 
```

```python
df_total= pd.concat([df_1940,df_1950,df_1960,df_1970,df_1980,df_1990,df_2000,df_2010],axis=0)

df_total
```

```python
df_total_2=df_total.sort_values(by=['año'],ascending=True,)
df_total_2
```

```python



plt.plot(df_total_2['año'], df_total_2['duracion_minutos'], color='green')


plt.xlabel('Año')
plt.ylabel('Duración en Minutos')
plt.title('Relación entre Duración y Año')


plt.show()
```

```python
plt.plot(df_total_2['explicito'], df_total_2['año'], color='green')


plt.xlabel('Año')
plt.ylabel('Duración en Minutos')
plt.title('Relación entre Duración y Año')


plt.show()
```
